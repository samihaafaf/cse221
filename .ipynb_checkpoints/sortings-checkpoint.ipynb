```python
#insertion sort new type
ar = [3,2,44,21,3]
my = [2,3,44,5,890,331,8]
def insert(ar):
    #swap = 0
    for i in range(1,len(ar)):
        current = ar[i]
        j = i-1
        while j>=0 and ar[j]>current:
            ar[j+1] = ar[j]
            #swap+=1
            j-=1
        ar[j+1] = current
    #print(swap)
    return ar
def selsort(ar):
    swap = 0
    comp = 0
    for a in range(len(ar)-1,1,-1):
        last = a
        maxv = ar[a]
        maxind = a
        for b in range(last):
            comp+=1
            if ar[b]>maxv:
                maxv = ar[b]
                maxind = b
        temp = ar[maxind]
        ar[maxind] = ar[last]
        ar[last] = temp
        swap+=1
    print('swap sel',swap,'comp',comp)
    return ar
print(insert(my))
print(selsort(my))
```

```python
#quicksort mine
myAr = [60,20,50,30,12,34]
def partition(ar):
    pivot = ar[0]
    print('pivot val',pivot)
    i = 0
    j = i+1
    while j<len(ar):
        if ar[j]<pivot:
            i+=1
            temp = ar[i]
            ar[i] = ar[j]
            ar[j] = temp
        j+=1
    temp = ar[i]
    ar[i] = pivot
    ar[0] = temp
    return i
def quicksort(ar):
    if len(ar)==1:
        return ar
    else:
        x = partition(ar)
        print('partition is',x)
        ar1 = ar[:x]
        ar2 = ar[x:]
        if len(ar1)>0 and len(ar2)>0:
            ar1 = quicksort(ar1)
            ar2 = quicksort(ar2)
        return ar1+ar2
store = quicksort(myAr)  #sorted array
#print(partition(store))
#print(store)
#quicksort(myAr)
print(store)     #sorted array
print(myAr)
```

```python
myAr = [60,20,50,30,12,34]
def quicksort(ar):
    if len(ar)==1:
        return ar
    else:
        greater = []
        smaller = []
        pivot = ar[0]
        for a in range(1,len(ar)-1):
            if ar[a]>=pivot:
                greater.append(ar[a])
            else:
                smaller.append(ar[a])
        return quicksort(smaller)+pivot+quicksort(greater)

quicksort(myAr)
print(myAr)
```
